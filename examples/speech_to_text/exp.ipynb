```python
import pandas as pd
import os
```

```python
def extract_common_metrics(dir, metric_names):

    available_metrics = set()
    files = os.listdir(dir)

    result_files = [f for f in files if f.startswith("results_")]

    for result_file in result_files:
        file_path = os.path.join(dir, result_file)
        with open(file_path, "r") as f:
            contents = [line.strip() for line in f.read().split("\n") if line.strip()]            

            if not contents:
                continue
        header = contents[0].split()
        if not header:
                continue
        available_metrics.update(header)
        common_metrics = set(metric_names).intersection(available_metrics)
    if common_metrics:
        return common_metrics
    else:
        raise ValueError("The specified metrics are not available in the results")
```

```python
def process_simuleval_results(dir, metric_names):
    if not os.path.exists(dir):
        raise ValueError("Directory does not exist")
    
    files = os.listdir(dir)

    result_files = [f for f in files if f.startswith("results_")]

    common_metrics = extract_common_metrics(dir, metric_names)


    data = []
    for result_file in result_files:
        file_path = os.path.join(dir, result_file)
        with open(file_path, "r") as f:
            contents = f.read().split("\n")
            if not contents:
                continue
        data.append(contents)

    data = [[val.split() for val in row if val] for row in data]

    # Extract unique headers from the first column
    headers = list(set([item for sublist in data for item in sublist[0]]))
    df = pd.DataFrame(columns=headers)

    for row in data:
        values = dict(zip(row[0], row[1]))
        df = pd.concat([df, pd.DataFrame([values])], ignore_index=True)
        
    df = df.fillna(0.0)

    filtered_df = df[df.columns[df.columns.isin(common_metrics)]]

    if len(common_metrics) == 1:
        metric_name = list(common_metrics)[0]
        filtered_df = filtered_df[filtered_df[metric_name] != 0.0]

    return filtered_df
```

```python
metric_names = ['STR']
dir = "results"
```
